```
def plot_2_images(imgpath1, imgpath2, title):
  img_raw1 = cv2.imread(imgpath1)
  img_raw2 = cv2.imread(imgpath2)
  fig, (ax1, ax2) = plt.subplots(1, 2)
  fig.suptitle(title, y= 0.83)
  ax1.imshow(img_raw1)
  ax1.set_axis_off()
  ax2.imshow(img_raw2)
  ax2.set_axis_off()
  plt.show()
```
